```python
import pandas as pd
from datetime import datetime, timedelta
import numpy as np
```

#Mexendo com o dataset de pluviometria

```python
#Puzando o xlsx
df_pluviometria_01_2023 = pd.read_excel('2023-01-PLUVIOMETRIA-CGESP.xlsx', header=None, engine='openpyxl')
```

```python
#Remover as 5 primeiras linhas
df_pluviometria_01_2023 = df_pluviometria_01_2023.iloc[6:].reset_index(drop=True)
```

```python
#Renomear o índice das colunas
df_pluviometria_01_2023.columns = df_pluviometria_01_2023.iloc[0]
```

```python
#Remover a primeira linha, que agora é redundante
df_pluviometria_01_2023 = df_pluviometria_01_2023[1:].reset_index(drop=True)
```

```python
#Remover linhas de média regional e total
df_pluviometria_01_2023 = df_pluviometria_01_2023.drop([7, 20, 21, 22, 27, 37, 38]).reset_index(drop=True)
```

```python
#Renomear coluna dos bairros
df_pluviometria_01_2023 = df_pluviometria_01_2023.rename(columns={df_pluviometria_01_2023.columns[0]: 'Bairro'})
```

```python
#Remover a última coluna
ultima_coluna = df_pluviometria_01_2023.columns[-1]
df_pluviometria_01_2023 = df_pluviometria_01_2023.drop(columns=[ultima_coluna])
```

```python
#Renomear a linha da ZC para Sé
df_pluviometria_01_2023.at[19, 'Bairro'] = 'SE - Sé'
```

```python
#Renomeando a coluna dos bairros para siglas
df_pluviometria_01_2023['Bairro'] = df_pluviometria_01_2023['Bairro'].apply(lambda x: x[:2])
```

```python
#Alterando o valor do índice das colunas para o formato datetime

# Definir as datas inicial e final
data_inicial = datetime(2023, 1, 1)
data_final = datetime(2023, 1, 31)

# Lista para armazenar as datas
datas = []

# Gerar datas de 01/09/2023 a 30/09/2023
data_atual = data_inicial
while data_atual <= data_final:
    # Adicionar a data formatada à lista
    datas.append(data_atual.strftime('%Y-%m-%d'))
    # Avançar para o próximo dia
    data_atual += timedelta(days=1)

#Adicionando os nomes com datas ao índice de colunas
novos_nomes = [df_pluviometria_01_2023.columns[0]] + datas
df_pluviometria_01_2023.columns = novos_nomes
```

```python
#Transformando as datas em linhas combinadas com os bairros
df_pluviometria_01_2023 = pd.melt(df_pluviometria_01_2023, id_vars=['Bairro'], var_name='Data', value_name='Pluviosidade')
```

```python
#Reformatando a data
df_pluviometria_01_2023['Data'] = pd.to_datetime(df_pluviometria_01_2023['Data'], format='%Y-%m-%d')
```

```python
#Display
df_pluviometria_01_2023.info()
df_pluviometria_01_2023
```

#Mexendo com o dataset de alagamento

```python
#Puxando o csv
df_alagamentos_2023 = pd.read_csv('alagamento2023.csv')
```

```python
#Reformatando a data
df_alagamentos_2023['Data'] = pd.to_datetime(df_alagamentos_2023['Data'], format='%Y-%m-%d')
```

```python
#Renomeando a coluna dos bairros para siglas
siglasBairros = {'Aricanduva/ Vila Formosa':'AF', 'Butantã':'BT', 'Campo Limpo':'CL', 'Capela do Socorro':'CS', 'Casa Verde':'CV', 'Cidade Ademar':'AD', 'Cidade Tiradentes':'CT', 'Ermelino Matarazzo':'EM', 'Freguesia do Ó':'FO', 'Guaianases':'GU', 'Ipiranga':'IP', 'Itaim Paulista':'IT', 'Itaquera':'IQ', 'Jabaquara':'JA', 'Jaçanã/ Tremembé':'JT', 'Lapa':'LA', "M'Boi Mirim":'MB', 'Mooca':'MO', 'Parelheiros':'PA', 'Penha':'PE', 'Perus':'PR', 'Pinheiros':'PI', 'Pirituba/Jaraguá':'PJ', 'Santana':'ST', 'Santo Amaro':'SA', 'Sapopemba':'SB', 'São Mateus':'SM', 'São Miguel Paulista':'MP', 'Sé':'SE', 'Vila Maria/ Vila Guilherme':'MG', 'Vila Mariana':'VM', 'Vila Prudente':'VP'}
df_alagamentos_2023['Bairro'] = df_alagamentos_2023['Bairro'].replace(siglasBairros)
```

```python
#Renomeando a coluna das zonas para siglas
siglasZonas = {'Zona Leste':'ZL', 'Zona Norte':'ZN', 'Centro':'ZC', 'Zona Oeste':'ZO', 'Zona Sul':'ZS', 'Zona Sudeste':'ZL'}
df_alagamentos_2023['Zona'] = df_alagamentos_2023['Zona'].replace(siglasZonas)
```

```python
#Colocar o Ipiranga na zona sul
df_alagamentos_2023['Zona'] = np.where(df_alagamentos_2023['Bairro'] == 'IP', 'ZS', df_alagamentos_2023['Zona'])
```
